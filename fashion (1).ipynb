```python
#Show product sales where bubble size represents revenue 
```

```python
np.random.seed(42)
prices = np.random.uniform(10, 500, size=200)  # 200 random product prices
quantities = np.random.poisson(lam=20, size=200)  # Sales quantity following Poisson distribution
revenues = prices * quantities  # Calculate revenue for each product (price * quantity)
```

```python
bubble_sizes = np.sqrt(revenues) * 5
```

```python
plt.scatter(prices, quantities, s=bubble_sizes, color='grey', alpha=0.6, edgecolors='black')
plt.title('Product Sales with Bubble Size Representing Revenue')
plt.xlabel('Product Price')
plt.ylabel('Sales Quantity')
plt.show()
```

```python
#Display the cumulative sales revenue over time
```

```python
months = pd.date_range(start='2024-01-01', periods=12, freq='M')
prices = np.random.uniform(10, 500, size=12)  #random product prices
quantities = np.random.poisson(lam=20, size=12)  #random sales quantities using Poisson distribution
```

```python
revenues = prices * quantities
cumulative_revenue = np.cumsum(revenues)
```

```python
plt.fill_between(months, cumulative_revenue, color='pink', alpha=0.6, label='Cumulative Revenue')
plt.title('Cumulative Sales Revenue Over Time')
plt.xlabel('Month')
plt.ylabel('Cumulative Revenue')
plt.show()
```
